```python
def open_input(day='01', suffix="input"):
    with open(f"{day}-{suffix}", "r") as file:
        lines = file.readlines()
    data_raw = [line.replace("\n", "") for line in lines]
    data_raw = "\n".join(data_raw)
    return data_raw

input_raw = open_input(day='01')
input_raw
```

```python
input_test_raw = open_input(day='01', suffix="test")
input_test_raw
```

```python
from ast import Tuple


def preprocess_data (data):
    # dtype='U10'
    rotations = []
    for row in data.split("\n"):   
        rotations.append((row[0], row[1:]))

    return rotations
test_data = preprocess_data(input_test_raw)
input_data = preprocess_data(input_raw)
display(test_data)
```

```python
def solution(data):
    solution = 0
    
    start = 50
    for direction, steps in data:
        rounds = int(steps) // 100
        solution += rounds
        
        if direction == 'L':
            start_next = (start - int(steps)) % 100
            if start != 0 and start_next > start or start_next == 0:
                solution += 1
            
        elif direction == 'R':
            start_next = (start + int(steps)) % 100
            if start != 0 and start_next < start or start_next == 0:
                solution += 1
        start = start_next
    return solution

sol_input = solution(input_data)
sol_test = solution(test_data)
print("Test Solution:", sol_test)
print("Input Solution:", sol_input)
```
