```python
def open_input(day='07', suffix="input"):
    with open(f"{day}-{suffix}", "r") as file:
        lines = file.readlines()
    data_raw = [line.replace("\n", "") for line in lines]
    data_raw = "\n".join(data_raw)
    return data_raw

input_raw = open_input()
input_raw
```

```python
input_test_raw = open_input( suffix="test")
input_test_raw
```

```python
def preprocess_data (data):
    # dtype='U10'
    ranges = []
    for row in data.split("\n"):   
        ranges.append([a for a in row])
    
    return np.array(ranges)
test_data = preprocess_data(input_test_raw)
input_data = preprocess_data(input_raw)
display(test_data)
```

```python
def solution(data):
    solution = 0
    
    h, w = data.shape
    start = np.where(data[0] == 'S')[0].item()
    last_beams = [start]

    for row in range(1, h):
        next_beams = []
        for beam in last_beams:
            if data[row][beam] == '.' and beam not in next_beams:
                next_beams.append(beam)
            elif data[row][beam] == '^':
                
                solution += 1
                if beam -1 >= 0 and beam - 1 not in next_beams:
                    next_beams.append(beam - 1)
                if beam + 1 < w and beam + 1 not in next_beams:
                    next_beams.append(beam + 1)
                    
        last_beams = next_beams
            
    return solution

sol_test = solution(test_data)
print("Test Solution:", sol_test)
sol_input = solution(input_data)
print("Input Solution:", sol_input)
```

```python


def solution2(data):
    solution = 1
    
    h, w = data.shape
    start = np.where(data[0] == 'S')[0].item()
    last_beams = {start: 1}

    for row in range(1, h):
        next_beams = defaultdict(int)
        for beam in last_beams.keys():
            if data[row][beam] == '.':
                next_beams[beam] += last_beams[beam]
            elif data[row][beam] == '^':
                
                solution += last_beams[beam] 
                if beam -1 >= 0:
                    next_beams[beam - 1] += last_beams[beam]
                if beam + 1 < w:
                    next_beams[beam + 1] += last_beams[beam]
                    
        last_beams = next_beams
        
    return solution

sol_test = solution2(test_data)
print("Test Solution:", sol_test)
sol_input = solution2(input_data)
print("Input Solution:", sol_input)
```
