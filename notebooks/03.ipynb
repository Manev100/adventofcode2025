```python
def open_input(day='03', suffix="input"):
    with open(f"{day}-{suffix}", "r") as file:
        lines = file.readlines()
    data_raw = [line.replace("\n", "") for line in lines]
    data_raw = "\n".join(data_raw)
    return data_raw

input_raw = open_input()
input_raw
```

```python
input_test_raw = open_input( suffix="test")
input_test_raw
```

```python
from ast import Tuple


def preprocess_data (data):
    # dtype='U10'
    rotations = []
    for row in data.split("\n"):   
        rotations.append([int(a) for a in row])

    return rotations
test_data = preprocess_data(input_test_raw)
input_data = preprocess_data(input_raw)
display(test_data)
```

```python
def solution(data):
    solution = 0
    
    for bank in data:
        first = bank[:-1]
        largest_idx = np.argmax(np.array(first))
        second = bank[largest_idx+1:]
        largest_idx_2 = np.argmax(np.array(second))
        
        solution += first[largest_idx]*10 + second[largest_idx_2]
            
    return solution

sol_test = solution(test_data)
print("Test Solution:", sol_test)
sol_input = solution(input_data)
print("Input Solution:", sol_input)
```

```python
def solution2(data):
    solution = 0
    
    for bank in data:
        batteries = np.array(bank)
        for level in range(12):
            limit = -11+level
            idx = np.argmax(batteries[:limit]) if limit != 0 else np.argmax(batteries)
            solution += batteries[idx] * (10**(11-level))
            
            batteries = batteries[idx+1:]
        
            
    return solution

sol_test = solution2(test_data)
print("Test Solution:", sol_test)
sol_input = solution2(input_data)
print("Input Solution:", sol_input)
```
