```python
def open_input(day='05', suffix="input"):
    with open(f"{day}-{suffix}", "r") as file:
        lines = file.readlines()
    data_raw = [line.replace("\n", "") for line in lines]
    data_raw = "\n".join(data_raw)
    return data_raw

input_raw = open_input()
input_raw
```

```python
input_test_raw = open_input( suffix="test")
input_test_raw
```

```python
from ast import Tuple


def preprocess_data (data):
    # dtype='U10'
    ranges = []
    fresh, ids = data.split("\n\n")
    for row in fresh.split("\n"):   
        ranges.append([int(a) for a in row.split("-")])
        
    ids = [int(x) for x in ids.split("\n")]

    return ranges, ids
test_data = preprocess_data(input_test_raw)
input_data = preprocess_data(input_raw)
display(test_data)
```

```python
def solution(data):
    ranges, ids = data
    solution = 0
    
    for id in ids:
        for range in ranges:
            if range[0] <= id <= range[1]:
                solution += 1
                break
        
    
    return solution

sol_test = solution(test_data)
print("Test Solution:", sol_test)
sol_input = solution(input_data)
print("Input Solution:", sol_input)
```

```python
def solution2(data):
    ranges, ids = data
    solution = 0
    
    non_overlapping_ranges = []
    
    def check_if_overlapping(range1, range2):
        if not (range1[0] < range2[0] and range1[1] < range2[0]) and not (range1[0] > range2[1] and range1[1] > range2[1]):
            return True
        else:
            return False
    
    for range in ranges:
        current_range = range.copy()
        next_non_overlapping = []
        for other_range in non_overlapping_ranges:
            if check_if_overlapping(current_range, other_range):
                current_range = [min(current_range[0], other_range[0]), max(current_range[1], other_range[1])]
            else:
                next_non_overlapping.append(other_range)
        
        non_overlapping_ranges = next_non_overlapping
        non_overlapping_ranges.append(current_range)
    
    
    for range in non_overlapping_ranges:
        solution += range[1] - range[0] + 1
        
    return solution

sol_test = solution2(test_data)
print("Test Solution:", sol_test)
sol_input = solution2(input_data)
print("Input Solution:", sol_input)
```
